```python
def perf_measure(y_actual, y_hat):
    """code heavily inspired on: https://stackoverflow.com/questions/31324218/scikit-learn-how-to-obtain-true-positive-true-negative-false-positive-and-fal"""
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]:
           TP += 1
        if y_hat[i] and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]!=y_hat[i]:
           TN += 1
        if y_hat[i] and y_actual[i]!=y_hat[i]:
           FN += 1
    return(TP, FP, TN, FN)

def calculate_precision(TP, FP):
    return TP / (TP + FP)

def calculate_accuracy(TP, TN, FP, FN):
    return (TP + TN)/(TP + TN + FP + FN)
    
    
def calculate_behaviours(data):
    precision_behav = []
    accuracy_behav = []
    for i in range(1,9):
        choice = list(data['firstchoice_'+str(i)])
        intended = list(data['intended_'+str(i)])
        TP, FP, TN, FN = perf_measure(intended, choice)
        accuracy_behav.append(calculate_accuracy(TP, TN, FP, FN))
        precision_behav.append(calculate_precision(TP,FP))
    return precision_behav, accuracy_behav
```
